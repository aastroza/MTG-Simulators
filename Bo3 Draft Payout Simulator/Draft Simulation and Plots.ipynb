```python
import pandas as pd
from mtga_bo3_draft_simulator import generate_prize_distributions
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = generate_prize_distributions()
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df['Winrate'],df['Gems-PRE'], label='Gems Premier')
plt.plot(df['Winrate'],df['Gems-TRA'], label='Gems Traditional')
plt.plot(df['Winrate'],df['NetGemsPRE'], label='NetGems Premier')
plt.plot(df['Winrate'],df['NetGemsTRA'], label='NetGems Traditional')
plt.xlim(0.2,1.01)
plt.legend()
plt.title("EV for Arena Bo3 Drafts")
plt.xlabel("WinRate")
plt.ylabel("EV (Gems)");
plt.show()
```

```python
df.to_csv('ArenaBo3drafts_data.csv')
```
